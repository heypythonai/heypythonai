#8-1 모듈

```python
def user_age(year):
  return 2024-year

def user_gender(minnum):
  if int(minnum[7])%2==0:
    return '여성'
  else:
    return '남성'
```

```python
import heypython
age = heypython.user_age(1999)
gender = heypython.user_gender('790202-1664019')
print(age)
print(gender)
```

```python
import heypython as hp
age = hp.user_age(1999)
gender = hp.user_gender('790202-1664019')
print(age)
print(gender)
```
