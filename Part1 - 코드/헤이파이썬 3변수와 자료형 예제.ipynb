#3-10 집합자료형

```python
price = [100,200,300,100,200,300]
set(price)
```

```python
price1 = set([100,200,300,400])
price2 = set([10,200,150,400])
result1 = price1 & price2
result2 = price1.intersection(price2)
print(result1)
print(result2)
```

```python
price1 = set([100,200,300,400])
price2 = set([10,200,150,400])
result1 = price1 | price2
result2 = price1.union(price2)
print(result1)
print(result2)
```

```python
price1 = set([100,200,300,400])
price2 = set([10,200,150,400])
result1 = price1 - price2
result2 = price1.difference(price2)
print(result1)
print(result2)
```
