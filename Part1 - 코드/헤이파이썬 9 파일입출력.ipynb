#9-1 사용자입출력

```python
name = input("이름을 입력하세요: ")
print(name)
```

```python
age = int(input("나이를 입력하세요: "))
print(age)
```

```python
name, age, phone = input("이름과 나이, 핸드폰 번호를 빈칸으로 구분해서 입력해 주세요: ").split()
print(name, age, phone)
```

```python
name = "김마트"
age = 25
phone = "010-0000-0000"
print(name)
print(age)
print(phone)
print(name, age, phone)
```

```python
name = "김마트"
phone = "010-0000-0000"
print(name+" : " + phone)
```

```python
name = "김마트"
age = 25
phone = "010-0000-0000"
print(name, age, phone, sep=', ')
print(name, age, phone, sep=' / ')
```

```python
name = "김마트"
age = 25
phone = "010-0000-0000"
print(name)
print(age)
print(phone)
print(name, end = ' ')
print(age, end = ', ')
print(phone, end = ' ')
```

#9-2 파일읽고쓰기

```python
file = open("새파일.txt", "w", encoding='utf8')
file.close()
```

```python
file = open("가격표.txt", "w", encoding="utf8")
file.write("콜라 : 1600원\n") # 가격표.txt 파일에 "콜라 : 1600원" 쓰기
file.write("사이다 : 1500원") # 가격표.txt 파일에 "사이 : 1500원" 쓰기
file.close()
```

```python
file = open("가격표.txt", 'r')
print(file)
result = file.read()
print(result)
```

```python
file = open("가격표.txt", 'r')
while 1:
  line = file.readline()
  if not line:
    break
  print(line)
```

```python
file = open("가격표.txt", 'r')
lines = file.readlines()
print(lines)
for line in lines:
  print(line)
```

```python
file = open("가격표.txt", 'a')
file.write("생수 : 1000원")
file.close()
file = open("가격표.txt", 'r')
lines = file.readlines()
for i in lines:
  print(i)
```

```python
with open("가격표_with 예제.txt", "w") as file:
  file.write("콜라 : 1600원\n")
  file.write("사이다 : 1500원\n")
with open("가격표_with 예제.txt", "r") as file:
  print(file.read())
```
